# MULTISPECTRAL IMAGING

## What is Multispectral Imaging?
Multispectral imaging captures image data within specific wavelength ranges across the electromagnetic spectrum (e.g., visible light, near-infrared), rather than just spatial resolution levels.

- Why it is important:
    - It allows for the extraction of information that the human eye cannot see, such as chemical composition or internal quality.
    - It enables the analysis of spectral signatures to identify materials, assess plant health, or detect defects (e.g., sugar content or maturity in fruits).

- Examples:
    - Precision Agriculture (e.g., crop monitoring, fruit quality analysis)
    - Remote Sensing (Satellite imagery for land use)
    - Food Safety & Quality Inspection

## What is a Band?
- Each band in an image represents a different wavelength range of the electromagnetic spectrum.
    ### Why it is important:
    - Different materials have different reflection properties at different wavelengths, allowing materials to be distinguished.
- Examples:
    - Red, green, blue bands (RGB)
    - Infrared bands
    - Ultraviolet bands

    ### 37-Band Multispectral Images

    - Application Areas:
        * Agriculture

        * City planning
    ### Uses of Multispectral Images
     - Remote Sensing:
       * Monitoring of land use changes
       * Evaluation of natural disasters
       * Agricultural productivity analyses
     - Medicine

       * Brain imaging
     - Military Applications:

## NDVI, EVI, SAVI Indices
- These indices are vector values (vegetation cover index) used to assess the health and density of vegetation.

    ### NDVI (Normalised Difference Vegetation Index):

    - Uses the difference between the infrared and red bands.
    - It is used to determine the presence and density of vegetation.
    ### EVI (Enhanced Vegetation Index):

    - A more sensitive index than NDVI.
    - It reduces soil effects.
    ### SAVI (Soil-Adjusted Vegetation Index):

    - Used to further reduce soil impacts.



    - Areas of Use:
    * Drought monitoring
    * Estimation of plague spread
    * Agricultural productivity analyses

```python
import os
import numpy as np
import matplotlib.pyplot as plt


file_path = r"multispectrallmagesArrays\Crimson_37bands_4.npy"


def calculate_ndvi(nir, red):
    return (nir - red) / (nir + red + 1e-6)

def calculate_evi(nir, red, blue):
    
    denominator = nir + 6 * red - 7.5 * blue + 1
    denominator[denominator == 0] = 1e-6  # Sıfır olanları küçük bir sayı ile değiştir
    return 2.5 * (nir - red) / denominator

def calculate_savi(nir, red, L=0.5):
    return (nir - red) / (nir + red + L) * (1 + L)


data = np.load(file_path)

# Bantları seç (örnek: NIR = 25, RED = 12, BLUE = 3)
nir_band = data[:, :, 25]
red_band = data[:, :, 12]
blue_band = data[:, :, 3]


ndvi = calculate_ndvi(nir_band, red_band)
evi = calculate_evi(nir_band, red_band, blue_band)
savi = calculate_savi(nir_band, red_band)

# Geçersiz değerleri sıfıra değiştir (NaN değerler için)
evi = np.nan_to_num(evi, nan=0)


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(ndvi, cmap='RdYlGn')
plt.title("NDVI")
plt.colorbar()

plt.subplot(1, 3, 2)
plt.imshow(evi, cmap='RdYlGn')
plt.title("EVI")
plt.colorbar()

plt.subplot(1, 3, 3)
plt.imshow(savi, cmap='RdYlGn')
plt.title("SAVI")
plt.colorbar()

plt.show()
data.shape
```

#### Project objective :
    My project aim is to see the effect of different indices on model accuracy when training a model using Multispectral image.   

### Dataset :

        Resolution: Resized to 140x200 pixels.
        Segmentation: Images were automatically preprocessed and denoised.
        Augmentation: Data augmentation was applied with the Albumentations library to increase the training data.

        

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

def load_data(data_dir):
    data = []
    labels = []
    
    for file_name in os.listdir(data_dir):
        if file_name.endswith(".npy"):
            file_path = os.path.join(data_dir, file_name)
            # Etiketi dosya adından alıyoruz (örneğin, "Itum5_37bands_618.npy" => "Itum5")
            label = file_name.split("_")[0]
            
            # Veriyi yükle ve listeye ekle
            image_data = np.load(file_path)
            data.append(image_data)
            labels.append(label)

    return np.array(data), np.array(labels)


def calculate_evi(data):
    G = 2.5
    C1 = 6
    C2 = 7.5
    L = 10000
    
    band1 = data[:, :, :, 0]  # Kırmızı (Band1)
    band2 = data[:, :, :, 1]  # Mavi (Band2)
    band3 = data[:, :, :, 3]  # Kızılötesi (Band3)
    
    evi = G * ((band3 - band1) / (band3 + C1 * band1 - C2 * band2 + L))
    
    return np.mean(evi, axis=(1, 2))


def calculate_ndvi(data):
    band1 = data[:, :, :, 0]  # Kırmızı
    band37 = data[:, :, :, -1]  # Son bant
    ndvi = (band37 - band1) / (band37 + band1 + 1e-6)
    
    return np.mean(ndvi, axis=(1, 2))


def calculate_savi(data):
    L = 0.5
    band1 = data[:, :, :, 0]  # Kırmızı (Band1)
    band37 = data[:, :, :, -1]  # Son bant (Band37)
    savi = ((band37 - band1) / (band37 + band1 + 1e-6)) * (1.0 + L)
    
    return np.mean(savi, axis=(1, 2))


if __name__ == "__main__":
    
    data_dir = "multispectrallmagesArrays/"

    
    print("Veriler yükleniyor...")
    data, labels = load_data(data_dir)
    print(f"Toplam veri sayısı: {len(data)}")

    
    print("Özellikler hesaplanıyor...")
    evi_features = calculate_evi(data)
    ndvi_features = calculate_ndvi(data)
    savi_features = calculate_savi(data)

    
    X_train_evi, X_test_evi, y_train_evi, y_test_evi = train_test_split(evi_features.reshape(-1, 1), labels, test_size=0.2, random_state=42)
    X_train_ndvi, X_test_ndvi, y_train_ndvi, y_test_ndvi = train_test_split(ndvi_features.reshape(-1, 1), labels, test_size=0.2, random_state=42)
    X_train_savi, X_test_savi, y_train_savi, y_test_savi = train_test_split(savi_features.reshape(-1, 1), labels, test_size=0.2, random_state=42)

    
    print("Model eğitiliyor...")
    model = RandomForestClassifier(random_state=42)
    
    model.fit(X_train_evi, y_train_evi)
    y_pred_evi = model.predict(X_test_evi)
    
    model.fit(X_train_ndvi, y_train_ndvi)
    y_pred_ndvi = model.predict(X_test_ndvi)
    
    model.fit(X_train_savi, y_train_savi)
    y_pred_savi = model.predict(X_test_savi)

    
    print("Model değerlendiriliyor...")
    accuracy_evi = accuracy_score(y_test_evi, y_pred_evi)
    accuracy_ndvi = accuracy_score(y_test_ndvi, y_pred_ndvi)
    accuracy_savi = accuracy_score(y_test_savi, y_pred_savi)

    print(f"EVI Doğruluk: {accuracy_evi:.2f}")
    print(f"NDVI Doğruluk: {accuracy_ndvi:.2f}")
    print(f"SAVI Doğruluk: {accuracy_savi:.2f}")

   
    
    indices = ['EVI', 'NDVI', 'SAVI']
    accuracies = [accuracy_evi, accuracy_ndvi, accuracy_savi]

    plt.figure(figsize=(8, 6))
    plt.bar(indices, accuracies, color=['blue', 'green', 'red'])
    plt.title('Doğruluk Karşılaştırması')
    plt.xlabel('İndeksler')
    plt.ylabel('Doğruluk')
    plt.ylim(0, 1)
    plt.show()

    
    features_combined = np.stack([evi_features, ndvi_features, savi_features], axis=1)
    X_train_combined, X_test_combined, y_train_combined, y_test_combined = train_test_split(features_combined, labels, test_size=0.2, random_state=42)

    model.fit(X_train_combined, y_train_combined)
    y_pred_combined = model.predict(X_test_combined)

    accuracy_combined = accuracy_score(y_test_combined, y_pred_combined)
    print(f"Birleştirilmiş Özelliklerle Doğruluk: {accuracy_combined:.2f}")


    accuracies.append(accuracy_combined)
    indices.append('Birleştirilmiş')

    plt.figure(figsize=(8, 6))
    plt.bar(indices, accuracies, color=['blue', 'green', 'red', 'purple'])
    plt.title('Doğruluk Karşılaştırması (Birleştirilmiş Özelliklerle)')
    plt.xlabel('İndeksler')
    plt.ylabel('Doğruluk')
    plt.ylim(0, 1)
    plt.show()
```
